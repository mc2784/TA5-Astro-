```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
band = pd.read_csv('mag.csv')
band = band[25:48]
band
```

```python
mag = np.linspace(-30, 30)
band_ = pd.DataFrame(mag, columns = ['m'])
band_ = band_[23:39]
band_
```

```python
band_['m'] = band_['m'].astype(float)
def tex_R():
    a = 48000
    b = (1.51) * 10**(7)
    c = 3080
    d = 10**(((-1)*band_['m'])/2.5)
    e = math.pi
    f = (0.2/2)**2
    g = 0.23
    tex_R = a/(b*c*d*e*f*g)
    return tex_R
print(tex_R())
band_['tex_R'] = tex_R ()
band_    
```

```python
band_['tex_R'] = band_['tex_R'].astype(float)
band_['m'] = band_['m'].astype(float)

fig = plt.figure(figsize=(14,10))
ax = fig.add_subplot(1,1,1)
ax.plot(band_['tex_R'], band_['m'], '.b-', color = 'red', markersize = 1)
ax.set_xlabel('Time exposure / s')

plt.show()
```

```python
band_['m'] = band_['m'].astype(float)
def tex_G():
    a = 48000
    b = (1.51) * 10**(7)
    c = 3730
    d = 10**(((-1)*band_['m'])/2.5)
    e = math.pi
    f = (0.2/2)**2
    g = 0.14
    tex_G = a/(b*c*d*e*f*g)
    return tex_G
print(tex_G())
band_['tex_G'] = tex_G()
band_    
```

```python
band_['m'] = band_['m'].astype(float)
m = band_['m']
def tex_B(m):
    a = 48000
    b = (1.51) * 10**(7)
    c = 4260
    d = 10**(((-1)*m)/2.5)
    e = math.pi
    f = (0.2/2)**2
    g = 0.22
    m = band_['m']
    tex_B = a/(b*c*d*e*f*g)
    return tex_B
print(tex_B(m))
band_['tex_B'] = tex_B(m)
band_    
```

```python
def log_R():
    log_R = np.log10(band_['tex_R'])
    return log_R

def log_B():
    log_B = np.log10(band_['tex_B'])
    return log_B

def log_G():
    log_G = np.log10(band_['tex_G'])
    return log_G

band_['log_R'] = log_R()
band_['log_B'] = log_B()
band_['log_G'] = log_G()
band_    
```

```python
band_['tex_R'] = band_['tex_R'].astype(float)
band_['tex_G'] = band_['tex_G'].astype(float)
band_['tex_B'] = band_['tex_B'].astype(float)
band_['m'] = band_['m'].astype(float)

fig = plt.figure(figsize=(10,9))
ax = fig.add_subplot(1,1,1)
ax.plot(band_['m'], band_['tex_R'], '.b-', color = 'red', markersize = 1, label = 'R')
ax.plot(band_['m'], band_['tex_G'], '.b-', color = 'green', markersize = 1, label = 'G')
ax.plot(band_['m'], band_['tex_B'], '.b-', color = 'blue', markersize = 1, label = 'B')
ax.set_xlabel('m', fontsize = 23)
ax.set_ylabel('log(T$_{exp}$)', fontsize = 23)
plt.rcParams.update({'font.size':23})
plt.text(-2.5, 0.000001, 'Brighter', fontsize = 23)
plt.text(15.5, 0.000001, 'Dimmer', fontsize = 23)
ax.plot([10.97],[60], marker = '.', color = 'black',  ls='none', ms=20, label = 'RV Uma')
ax.plot([9.48],[30],marker = '.', color = 'teal',  ls='none', ms=20, label = 'RZ Cep')
ax.plot([6.4],[0.6],marker = '.', color = 'orange',  ls='none', ms=20, label = 'RZ Cas')
ax.set_yscale('log')

ax.legend()

plt.savefig("t_exp_graph.pdf", dpi = 1500)
plt.show()
```
